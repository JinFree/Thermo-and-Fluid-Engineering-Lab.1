### Python에서의 함수 작성 및 함수 활용 예시

````
def function(num_a, num_b):
    if num_a > 20:
        print("num_a is larger than 20")
        return num_a
    elif num_a == 10:
        return num_a * num_b
    else:
        result = 0
        for i in range(num_a):
            result = result + num_b
        return result

a = 10
b = 20
result = function(a, b)
print(result)
````
